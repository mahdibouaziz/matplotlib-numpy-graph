## EX

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## initial values for the begin + end + steps + nbPoint

```python
#initial values for the begin + end + steps
begin=0.5
end=1.5
nbPoint=100
steps=(end-begin)/nbPoint
```

## get the x_data (r in our exemples)

```python
x_data=np.arange(begin,end,steps)
```

## create our function FAD

```python
def f(x,epsilon):
    denum=math.sqrt((1-x**2)**2+(2*epsilon*x)**2)
    return 1/denum
```

## Create a function that plot our FAD

```python
def plotting(x_data,epsilon):
    y_data=np.arange(begin,end,steps)
    i=0
    for item in x_data:
        y_data[i]=f(x_data[i],epsilon)
        i+=1
    #Prepare our figure for Question1
    fig=plt.figure(1,figsize=(10,10))
    plt.plot(x_data,y_data)
```

## Question1

```python
epsilon=0
plotting(x_data,epsilon)
```

## Question2

```python
#Create an array for all epsilons
epsilons=[0.02,0.05,0.1]
for epsilon in epsilons:
    plotting(x_data,epsilon)
```
